```python
# FORMULAÇAO MATEMATICA DO PROBLEMA DE PL
# MIN Z = 6x1 + 10x2
# S.A.
# -x1 + x2 <= 2
# x1 <= 5
# x2 <= 6
# 3x1 + 5x2 >= 15
# 5x1 + 4x2 >= 20
# x1, x2 >= 0
```

```python
import matplotlib.pyplot as plt
import numpy as np

# delimitacao do plano cartesiano
x1 = np.linspace(0, 30, 1000)
x2 = np.linspace(0,30,1000)
fig, ax = plt.subplots()
fig.set_size_inches(14.7, 8.27)

# restricoes do problema

y0 = 5
plt.axvline(y0, color = 'g', label = '$x_1 \leq 5$') # x1 <= 5

y1 = 6
plt.axhline(y1, color = 'r', label = '$x_2 \leq 6$') # x2 <= 6

y2 = 2 + x1
plt.plot(x1, y2, label = '$-x_1 + x_2 \leq 2$') # -x1 + x2 <= 2

y3 = 15/5 - 3*x1/5
plt.plot(x1, y3, label = '$3x_1 + 5x_2 \geq 15$') # 3x1 + 5x2 >= 15

y4 = 20/4 - 5*x1/4
plt.plot(x1, y4, label = '$5x_1 + 4x_2 \geq 20$') # 5x1 + 4x2 >= 20

# ajuste da vizualizaçao gráfica
plt.xlim(0, 10)
plt.ylim(0,8)

# Titulo dos eixos do gráfico
plt.xlabel(r'Número de produtos P1($x_1$)')
plt.ylabel(r'Número de produtos P2($x_2$)')

plt.legend(bbox_to_anchor = (1.05, 1), loc = 2, borderaxespad = 0.)

# preenchimento da região factivel
y21 = np.minimum(y2,y1)
y43 = np.maximum(y4,y3)
y4321 = np.minimum(y21,y43)

plt.fill_between(x1, y21, y4321, where = x1 <=5,  color = 'gray', alpha = 0.25)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)


# Vetor gradiente
fx1 = 6
fx2 = 10

# Curvas de nível
ax.quiver(fx1, fx2, color = 'deeppink', scale_units = 'xy', scale = 4)
X1, X2 = np.meshgrid(x1, x2)
Z = fx1*X1 + fx2*X2
plt.contour(X1, X2, Z, 40)

# Vizualização
plt.show()

```
